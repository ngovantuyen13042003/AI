```python
import numpy as np
import pandas as pd


data = pd.read_csv("weight-height.csv")
print(data.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(data[data["Gender"] == "Male"]["Height"] ,data[data["Gender"] == "Male"]["Weight"], 10, "green",label="Male", marker="+" )
plt.scatter(data[data["Gender"] == "Female"]["Height"] ,data[data["Gender"] == "Female"]["Weight"], 10, "pink", label="Female" )

plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend(title='Gender')
plt.title('Weight vs. Height by Gender')
```

```python
#4 (2)
from sklearn.linear_model import LinearRegression
# chiều cao là biến đọc lập còn cân nặng là biến phụ thuộc
# mục đích tìm ra điểm cắt 

x = data["Height"].values
y = data["Weight"].values

reg = LinearRegression().fit(x.reshape(-1, 1), y) 
print(f'w1 = {reg.coef_[0]}') # in ra biến độc lập
print(f'w0 = {reg.intercept_}') # in ra điểm cắt của mô hình
```
