```python
import numpy as np 
import matplotlib.pyplot as plt

# định nghĩa hàm mô hình 
def loss(w) :
    return w**2 + 5*np.sin(w)

# định nghĩa hàm đạo hàm
def grad(w) :
    return 2*w + 5*np.cos(w)
# tạo một mảng 1 chiều w. có giá trị từ -5 -> 5 với step = 0,1
w = np.arange(-4, 8, 0.1)
los = loss(w)
#  vẽ đồ thị 
plt.plot(w, los)


w = -10
alpha = 0.1
eps = 0.0001

while True:
    w = w - alpha*grad(w)
    plt.plot(w, loss(w), "ro")
    plt.pause(0.5)
    if abs(grad(w) < eps) :
        break

plt.show()

# in kết quả
print("giá trị nhỏ nhất của hàm loss: ", loss(w))
print("giá trị nhỏ nhất đạt được khi w = ", w)





```
