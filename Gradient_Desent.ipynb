```python
import numpy as np 
import matplotlib.pyplot as plt

# nh ngha hm m hnh 
def loss(w) :
    return w**2 + 5*np.sin(w)

# nh ngha hm o hm
def grad(w) :
    return 2*w + 5*np.cos(w)
# to mt mng 1 chiu w. c gi tr t -5 -> 5 vi step = 0,1
w = np.arange(-4, 8, 0.1)
los = loss(w)
#  v  th 
plt.plot(w, los)


w = -10
alpha = 0.1
eps = 0.0001

while True:
    w = w - alpha*grad(w)
    plt.plot(w, loss(w), "ro")
    plt.pause(0.5)
    if abs(grad(w) < eps) :
        break

plt.show()

# in kt qu
print("gi tr nh nht ca hm loss: ", loss(w))
print("gi tr nh nht t c khi w = ", w)





```
