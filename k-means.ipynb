```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs # create data
from sklearn.cluster import KMeans # xay dung thuat toan k-means
```

```python
x,y = make_blobs(n_samples=200, centers=3, cluster_std=2.75, random_state=99)
```

```python
x
```

```python
x1 = x[:,0] # get all data of column 0
x2 = x[:,1] # get all data of column 1
```

```python
plt.scatter(x1,x2)
```

```python
kMeans = KMeans(n_clusters=3, random_state=99)
```

```python
clustering_output=kMeans.fit(x)
```

```python
clustering_output.labels_
```

```python
clustering_output.cluster_centers_
```

```python
plt.scatter(x1,x2,color="blue")
plt.scatter(clustering_output.cluster_centers_[:,0], clustering_output.cluster_centers_[:,1], color="red")
```

```python
color_dict = {0:'b', 1:'g', 2:'y'}
colors = [color_dict[i] for i in kMeans.labels_]
```

```python
plt.scatter(x1,x2,color=colors)
plt.scatter(clustering_output.cluster_centers_[:,0], clustering_output.cluster_centers_[:,1], color="red")
```

```python
input_test = np.array([4,7])
```

```python
testing_output = kMeans.predict(input_test.reshape(-1,2))
```

```python
testing_output[0]
```

```python
plt.scatter(x1,x2,color=colors)
plt.scatter(clustering_output.cluster_centers_[:,0], \
    clustering_output.cluster_centers_[:,1], color="red")
plt.scatter(input_test[0], input_test[1], color = color_dict[testing_output[0]], marker='*')
```

```python
plt.scatter(x1,x2,color=colors)
plt.scatter(clustering_output.cluster_centers_[:,0], \
    clustering_output.cluster_centers_[:,1], color="red")
plt.scatter(input_test[0], input_test[1], color = 'pink', marker='*')
```
